```python
# Importação e instalação de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar fonte de dados
dados = pd.read_csv('dados_itens_meli.csv', encoding='latin1')
```

```python
# Exibir as primeiras linhas do DataFrame
print(dados.head())
```

```python
# Obter informações sobre o DataFrame
print(dados.info())
```

```python
# Resumo estatístico das variáveis numéricas
print(dados.describe())
```

```python
# Contagem de valores únicos em cada coluna
print(dados.nunique())
```

```python
# Gráfico de dispersão entre Preço e Preço Base
plt.scatter(dados['price'], dados['base_price'])
plt.title('Gráfico de Dispersão entre Variáveis Numéricas')
plt.xlabel('Preço')
plt.ylabel('Preço Base')
plt.show()
```

```python
# Gráfico de barras Quantidade de Produtos por buying_mode
sns.countplot(x='buying_mode', data=dados)
plt.title('Quantidade de Produtos por Categoria')
plt.xlabel('Modo de Compra')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de barras Quantidade de Produtos por Categoria
sns.countplot(x='category_id', data=dados)
plt.title('Quantidade de Produtos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Contagem')
plt.show()
```
